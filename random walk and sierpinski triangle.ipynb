## Simulating a random walk with matplotlib

```python
import matplotlib.pyplot as plt
import choice from 
%matplotlib inline
plt.rc('figure', figsize = (12,7))

def rand_walk(step_num):
    walk = []
    step1 = choice([1,-1])
    walk.append(step1)
    
    for i in range(1,step_num):
        step_choice = choice([1, -1])
        next_step = walk[i - 1] + step_choice
        walk.append(next_step)
    
    return walk
    
    
```

```python
random_w = rand_walk(100)
```

```python
plt.plot(random_w)
plt.title('A random walk', size = 14)
plt.show
```

```python
def plot_walk(walk):
    plt.plot(walk)
    plt.title("A random walk", size = 14)
    plt.xlabel("Number of steps")
    plt.ylabel("Distance from Origin")
    plt.show
    
def plot_walks(num, step_num):
    label_list = list(range(1, num+1))
    for i in range(0,num):
        plt.plot(rand_walk(step_num), label = 'Plot number' + ' ' + str(label_list[i]))
        plt.title("A random walk")
        plt.xlabel("Number of steps")
        plt.ylabel("Distance from the origin")
        plt.legend(loc = "lower left")
    plt.show    
    
```

```python
plot_walks(5, 100)
```

## Plot of a Sierpinski Triangle

```python
# I will use 3 transformations for the triangle

def trans_1(p):
    x = p[0]
    y = p[1]     # here i create a function to take a list of x and y as arguement
    
    x1 = 0.5 * x
    y1 = 0.5 * y   # implementing first transformation 
    return x1,y1
    
def trans_2(p):
    x = p[0]
    y = p[1]
    
    x1 = (0.5 * x) + 0.5
    y1 = (0.5 * y) + 0.5
    return x1,y1
    
def trans_3(p):
    x = p[0]
    y = p[1]
    
    x1 = (0.5 * x) + 1
    y1 = 0.5 * y
    return x1,y1

from random import choice

transformations = [trans_1, trans_2, trans_3]
a1 = [0]
b1 = [0]
a,b = 0,0

for i in range(1000000):
    trans = choice(transformations)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('figure', figsize = (16,16))

plt.plot(a1,b1, 'o')
plt.show
plt.savefig('Sierpinski Triangle.png')
```
